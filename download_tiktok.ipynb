```python
not_saved_list = []
general_path = fr"C:\Users\31648\OneDrive - UvA\Documenten\KI\Project 4\Notebook"
wanted_path = fr"C:\Users\31648\OneDrive - UvA\Documenten\KI\Project 4\Notebook\All_TikToks_and_data\TikTokVids_IM"
file = fr"C:\Users\31648\OneDrive - UvA\Documenten\KI\Project 4\Notebook\All_TikToks_and_data\ALL_TIKTOK_irish_moss_19_12_2025.csv"
with open(file, "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    count = 0

    for row in reader:
        url = (row[0])
        count += 1
        try:
            result = pyk.save_tiktok(url, True, fr"C:\Users\31648\OneDrive - UvA\Documenten\KI\Project 4\Notebook\All_TikToks_and_data\TikToks_IM_data\IM_video_data{count}.csv", browser_name=None)
            print("This URL is saved", url)

            p = urlparse(url)
            parts = p.path.strip("/").split("/")
            username = parts[0]
            video_id = parts[2]
            old_name = f"{username}_video_{video_id}.mp4"

            old_video_path = os.path.join(general_path, old_name)
            new_video_path = os.path.join(wanted_path, f"IM_video{count}.mp4")

            os.rename(old_video_path, new_video_path)
        
        except:
            print("")
            print(f"This URL: {url} is NOT saved. Do this manually")
            not_saved_list.append([count, url])
            print("")

output_csv = fr"C:\Users\31648\OneDrive - UvA\Documenten\KI\Project 4\Notebook\All_TikToks_and_data\TikToks_IM_data\failed_downloads.csv"
with open(output_csv, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["count", "url"])
    writer.writerows(not_saved_list)
```
