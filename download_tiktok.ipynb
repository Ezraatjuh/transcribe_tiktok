```python
import numpy as py
import pyktok as pyk
import csv
import os
from urllib.parse import urlparse
```

```python
file = "The output file of V4tiktok_scroll_inf.ipynb will be this input file"
video_data_dir = "This will be the general name of the files where the data will be stored"
general_path = "General path directory where the videos will be saved"
wanted_path = "Path directory where you want the videos to be saved"
output_csv = "If a video could not be downloaded, it will be saved here"
```

```python
not_saved_list = []
with open(file, "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    count = 0

    for row in reader:
        url = (row[0])
        count += 1
        try:
            result = pyk.save_tiktok(url, True, f"{video_data_dir}{count}.csv", browser_name=None)
            print("This URL is saved", url)

            p = urlparse(url)
            parts = p.path.strip("/").split("/")
            username = parts[0]
            video_id = parts[2]
            old_name = f"{username}_video_{video_id}.mp4"

            old_video_path = os.path.join(general_path, old_name)
            new_video_path = os.path.join(wanted_path, f"IM_video{count}.mp4")

            os.rename(old_video_path, new_video_path)
        
        except:
            print("")
            print(f"This URL: {url} is NOT saved. Do this manually")
            not_saved_list.append([count, url])
            print("")

with open(output_csv, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["count", "url"])
    writer.writerows(not_saved_list)
```
